### Data Preparation

```python
# Transforming datetime objects
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
```

```python
# Labeling payment methods
payment_map = {
    1: "Credit Card",
    2: "Cash",
    3: "No Charge",
    4: "Dispute"
}

df["payment_label"] = df["payment_type"].map(payment_map)
```

```python
# Trip duration im minutes transformation
df["trip_duration_min"] = (
    df["tpep_dropoff_datetime"] - df["tpep_pickup_datetime"]
).dt.total_seconds() / 60
```

```python
# Cleaning data
df = df[df["trip_duration_min"] > 0] # Removing trips with no duration time
df = df.fillna(0) # Fill NaN values to "0"
```

#### Data Graphics

```python
# Trip distance distribution
df_distance = df[df["trip_distance"] < 25]
plt.hist(df_distance["trip_distance"], bins=25)
plt.title("Trip Distance Distribution ")
plt.xlabel("Distance (miles)")
plt.ylabel("Frequency")
plt.savefig("../Data/Silver/graph1_distance.png")
plt.clf()
```

```python
# Payment method distribution
counts = df["payment_label"].value_counts()
counts.plot(kind="bar")
plt.title("Payment Method Distribution")
plt.xlabel("Payment Type")
plt.ylabel("Number of Trips")
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig("../Data/Silver/graph2_payment_type.png")
plt.clf()
```

```python
# Daily revenue over time
df.groupby(df["tpep_pickup_datetime"].dt.to_period("D"))["total_amount"].sum().plot()
plt.title("Daily Revenue Over Time")
plt.xlabel("Date")
plt.ylabel("Total Revenue (dollar)")
plt.xticks(rotation=45)
plt.tight_layout() 
plt.savefig("../Data/Silver/graph3_revenue.png")
plt.clf()
```

```python
# Passenger count distribution
df_clean = df[(df["passenger_count"] > 0) & (df["passenger_count"] <= 6)]
counts = df_clean["passenger_count"].value_counts().sort_index()
counts.plot(kind="bar")
plt.title("Passenger Count Distribution")
plt.xlabel("Number of Passengers")
plt.ylabel("Number of Trips")
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig("../Data/Silver/graph4_passenger_count.png")
plt.clf()
```

```python
# Trip duration distribution
df_clean = df[
    (df["trip_duration_min"] > 0) & 
    (df["trip_duration_min"] < 300)  
]
plt.hist(df_clean["trip_duration_min"], bins=50)
plt.title("Trip Duration Distribution")
plt.xlabel("Duration (minutes)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.savefig("../Data/Silver/graph5_duration.png")
plt.clf()
```
